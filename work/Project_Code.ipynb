```python
stateEnrollment = pd.read_csv('../data/Group Project/Overall-Enrollment-Coverage-2017-2022.csv')
# clean data (Get rid of unused columns)
stateEnrollment = stateEnrollment[['State or Region', 'Fall 2017 IPEDS Universe', 'Fall 2018 IPEDS Universe', 'Fall 2019 IPEDS Universe', 'Fall 2020 IPEDS Universe', 'Fall 2021 IPEDS Universe']]
stateEnrollment
```

```python
overallChange = pd.DataFrame()
overallChange['State or Region'] = stateEnrollment['State or Region']
overallChange['2018_Enrollment_Change'] = ((stateEnrollment['Fall 2018 IPEDS Universe'] / stateEnrollment['Fall 2017 IPEDS Universe']) - 1) * 100
overallChange['2019_Enrollment_Change'] = ((stateEnrollment['Fall 2019 IPEDS Universe'] / stateEnrollment['Fall 2018 IPEDS Universe']) - 1) * 100
overallChange['2020_Enrollment_Change'] = ((stateEnrollment['Fall 2020 IPEDS Universe'] / stateEnrollment['Fall 2019 IPEDS Universe']) - 1) * 100
overallChange['2021_Enrollment_Change'] = ((stateEnrollment['Fall 2021 IPEDS Universe'] / stateEnrollment['Fall 2020 IPEDS Universe']) - 1) * 100
overallChange
```

```python
US_overall_change = overallChange.head(1)
States_overall_change = overallChange[overallChange['State or Region'] != 'United States']

graph = US_overall_change.plot.bar(ylabel="Percent Change in Enrollment", title='Percent Change in Enrollment for all US 2018-2021', figsize=(7,5))
graph = graph.set_xticklabels(US_overall_change['State or Region'])
```

```python
# Set up Data sets
publicEnrollment = pd.read_csv('../data/Group Project/Public-Enrollment-Coverage-2017-2022.csv')
privateEnrollment = pd.read_csv('../data/Group Project/Private-Enrollment-Coverage-2017-2022.csv')

# clean data (Get rid of unused columns)
publicEnrollment = publicEnrollment[['State or Region', 'Fall 2017 IPEDS Universe', 'Fall 2018 IPEDS Universe', 'Fall 2019 IPEDS Universe', 'Fall 2020 IPEDS Universe', 'Fall 2021 IPEDS Universe']]
privateEnrollment = privateEnrollment[['State or Region', 'Fall 2017 IPEDS Universe', 'Fall 2018 IPEDS Universe', 'Fall 2019 IPEDS Universe', 'Fall 2020 IPEDS Universe', 'Fall 2021 IPEDS Universe']]

# get rid of Wyoming because of incomplete data
publicEnrollment = publicEnrollment[publicEnrollment['State or Region'] != "Wyoming"]
privateEnrollment = privateEnrollment[privateEnrollment['State or Region'] != "Wyoming"]
```

```python
#Public school percent change
publicChange = pd.DataFrame()
publicChange['State or Region'] = publicEnrollment['State or Region']
publicChange['2018_Enrollment_Change'] = ((publicEnrollment['Fall 2018 IPEDS Universe'] / publicEnrollment['Fall 2017 IPEDS Universe']) - 1) * 100
publicChange['2019_Enrollment_Change'] = ((publicEnrollment['Fall 2019 IPEDS Universe'] / publicEnrollment['Fall 2018 IPEDS Universe']) - 1) * 100
publicChange['2020_Enrollment_Change'] = ((publicEnrollment['Fall 2020 IPEDS Universe'] / publicEnrollment['Fall 2019 IPEDS Universe']) - 1) * 100
publicChange['2021_Enrollment_Change'] = ((publicEnrollment['Fall 2021 IPEDS Universe'] / publicEnrollment['Fall 2020 IPEDS Universe']) - 1) * 100

# seperate US from States
US_public_enrollment = publicChange.head(1)
States_public_enrollment = publicChange[publicChange['State or Region'] != "United States"]

US_public_enrollment
```

```python
#SMU and Peer Data
```

```python
peerUniversities = pd.read_csv('../data/Group Project/SMU and Peer Universities 2017-2022.csv')
peerUniversities
```

```python
# How did the number of applicants change from 2018-2022
appChanges = pd.DataFrame()
appChanges['University'] = peerUniversities['University']
appChanges['2018_Application_Change'] = ((peerUniversities['Applicant_Num_18'] / peerUniversities['Applicant_Num_17']) - 1) * 100
appChanges['2019_Application_Change'] = ((peerUniversities['Applicant_Num_19'] / peerUniversities['Applicant_Num_18']) - 1) * 100
appChanges['2020_Application_Change'] = ((peerUniversities['Applicant_Num_20'] / peerUniversities['Applicant_Num_19']) - 1) * 100
appChanges['2021_Application_Change'] = ((peerUniversities['Applicant_Num_21'] / peerUniversities['Applicant_Num_20']) - 1) * 100
appChanges['2022_Application_Change'] = ((peerUniversities['Applicant_Num_ 22'] / peerUniversities['Applicant_Num_21']) - 1) * 100
appChanges
```

```python
# Did the school admit more or less people?
admissionChanges = pd.DataFrame()
admissionChanges['University'] = peerUniversities['University']
admissionChanges['2018_Admission_Change'] = ((peerUniversities['Admitted_Num_18'] / peerUniversities ['Admitted_Num_17']) - 1) * 100
admissionChanges['2019_Admission_Change'] = ((peerUniversities['Admitted_Num_19'] / peerUniversities['Admitted_Num_18']) - 1) * 100
admissionChanges['2020_Admission_Change'] = ((peerUniversities['Admitted_Num_20'] / peerUniversities['Admitted_Num_19']) - 1) * 100
admissionChanges['2021_Admission_Change'] = ((peerUniversities['Admitted_Num_21'] / peerUniversities['Admitted_Num_20']) - 1) * 100
admissionChanges['2022_Admission_Change'] = ((peerUniversities['Admitted_Num_22'] / peerUniversities['Admitted_Num_21']) - 1) * 100
admissionChanges
```

```python
# What was the acceptance rate?
admissionRates = pd.DataFrame()
admissionRates['University'] = peerUniversities['University']
admissionRates['2018_Admission_Rate'] = (peerUniversities['Admitted_Num_18'] / peerUniversities ['Applicant_Num_18']) * 100
admissionRates['2019_Admission_Rate'] = (peerUniversities['Admitted_Num_19'] / peerUniversities ['Applicant_Num_19']) * 100
admissionRates['2020_Admission_Rate'] = (peerUniversities['Admitted_Num_20'] / peerUniversities ['Applicant_Num_20']) * 100
admissionRates['2021_Admission_Rate'] = (peerUniversities['Admitted_Num_21'] / peerUniversities ['Applicant_Num_21']) * 100
admissionRates['2022_Admission_Rate'] = (peerUniversities['Admitted_Num_22'] / peerUniversities ['Applicant_Num_ 22']) * 100
admissionRates
```

```python
# Of those admitted, how many decided to enroll?
decisionRates = pd.DataFrame()
decisionRates['University'] = peerUniversities['University']
decisionRates['2018_Decision_Rate'] = (peerUniversities['Total_Enrollment_18'] / peerUniversities['Admitted_Num_18']) * 100
decisionRates['2019_Decision_Rate'] = (peerUniversities['Total_Enrollment_19'] / peerUniversities['Admitted_Num_19']) * 100
decisionRates['2020_Decision_Rate'] = (peerUniversities['Total_Enrollment_20'] / peerUniversities['Admitted_Num_20']) * 100
decisionRates['2021_Decision_Rate'] = (peerUniversities['Total_Enrollment_21'] / peerUniversities['Admitted_Num_21']) * 100
decisionRates['2022_Decision_Rate'] = (peerUniversities['Total_Enrollment_22'] / peerUniversities['Admitted_Num_22']) * 100
decisionRates
```

```python
plot3 = appChanges.plot.bar(xlabel = "University", ylabel="Percent Change in Applications", title='Percent Change in Applications by School', figsize=(20,20))
plot3 = plot3.set_xticklabels(appChanges['University']);
```
