```python
import pandas as pd
allUniversities = pd.read_csv('../data/Group Project/SMU and Peer Universities 2017-2022.csv')
```

```python
applications = pd.DataFrame()
applications['University'] = allUniversities['University']
applications['2017_Applications'] = allUniversities['Applicant_Num_17']
applications['2018_Applications'] = allUniversities['Applicant_Num_18']
applications['2019_Applications'] = allUniversities['Applicant_Num_19']
applications['2020_Applications'] = allUniversities['Applicant_Num_20']
applications['2021_Applications'] = allUniversities['Applicant_Num_21']
applications['2022_Applications'] = allUniversities['Applicant_Num_ 22']                            
applications
```

```python
graph = applications.plot.bar(xlabel="University", ylabel="Number of Applicants", title='Number of Applicants per University from 2017-2022', figsize=(20,25), color=['orange','pink','red','blue','green','black'])
graph.set_xticklabels(applications['University'])
```

```python
graph.figure.savefig('Applications.pdf')
```
